## Problem 1

Stirling's Approximation for $ln(n!)$ is:

\begin{equation}
ln(n!) \approx nln(n) − n.
\end{equation}

Write a program to output $ln(n!)$ and its Stirling's approximation for $2 < n < 20$. Plot on the same graph the exact value and the approximate value as a function of n. Be sure to lable your curves and axes. **Hint**: the math module (and numpy.math) provides a factorial() method.

**Expected Plot (note this one has 2 < n < 100)**


```python
import math

def ln_function(i):
    n_values = np.arange(2, i)
    ln_factorial_exact = np.array([math.log(math.factorial(n)) for n in n_values])
    stirling_approx = n_values * np.log(n_values) - n_values
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, ln_factorial_exact, label='Exact ln(n!)', marker='H')
    plt.plot(n_values, stirling_approx, label="Stirling's Approximation", marker='x')
    plt.xlabel('n')
    plt.ylabel('y')
    plt.title("Stirling's Approximation")
    plt.legend()
    plt.grid(False)
    plt.show()
ln_function(20)
```

Now modify your program to output the exact value of $ln(n!)$, Stirling's approximation for it, and the relative error in the approximation for the first value of n where the releative error is less than 1%. The relative error in percent is defined as 100$\times$(exact-approx)/exact. Your output should quote the numbers to 3 decimal places. **Hint**: You may need to increase your n value and you may get an error message. Can you determine why you get an error and how to correctly make this comparison? 

```python
#found that the larger the n value the relative error stays around 1% but smaller n values it's harder to tell
#also had to redefine the function again because it was giving me a weird call error saying a line wasn't there but it was so idk
#otherwise I would've called the same function from earlier but I think this works


def ln_function_error(i):
    n_values = np.arange(2, i)
    ln_factorial_exact = np.array([math.log(math.factorial(n)) for n in n_values])
    stirling_approx = n_values * np.log(n_values) - n_values
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, ln_factorial_exact, label='Exact ln(n!)', marker='H')
    plt.plot(n_values, stirling_approx, label="Stirling's Approximation", marker='x')
    plt.xlabel('n')
    plt.ylabel('y')
    plt.title("Stirling's Approximation")
    plt.legend()
    plt.grid(False)
    plt.show()
    relative_error = 100*((ln_factorial_exact - stirling_approx)/(ln_factorial_exact))
    for i, error in enumerate(relative_error):
        if error < 1:
            n = n_values[i]
            print(f"For n = {n}, the relative error is less than 1%.")
            print(f"Exact value: {ln_factorial_exact[i]:.3f}")
            print(f"Stirling's approximation: {stirling_approx[i]:.3f}")
            print(f"Relative error: {error:.3f}%")
            break
    else:
        print("\nNo value of n within the range has a relative error less than 1%.")
ln_function_error(100)
```

## Problem 6

Using *linspace* Generate 100 equally spaced points from 0 to 1. Then plot the cube of the values generated from *linespace* vs the *linspace* generated values.

```python
x = np.linspace(0, 1, 100)
y = x ** 3
plt.figure(figsize=(5, 5))
plt.plot(x, y, label='$y = x^3$', color='blue', marker='')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y = x^3$ vs $x$')
plt.legend()
plt.grid(False)
plt.show()
```
